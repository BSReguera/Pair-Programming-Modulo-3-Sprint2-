```python
# Apertura y lectura del archivo csv
ventas_df = pd.read_csv('ventas.csv')
ventas_df
```

```python
productos_df = pd.read_csv('productos.csv', error_bad_lines=False, warn_bad_lines=False)
productos_df
```

```python
clientes_df = pd.read_csv('clientes.csv')
clientes_df
```

```python
lista_df = [ventas_df, productos_df, clientes_df]
```

```python
def exploracion (lista):
    """Esta función proporciona una descripción personalizada de un DataFrame,
    incluyendo estadísticas descriptivas y tipos de datos de cada columna.
    
    Argumentos:
    df : DataFrame de Pandas
        El DataFrame para el cual se generará la descripción
        
    La funcion no tiene return pero devuelve varios prints con
    la informacion que necesitamos:
    - describe separados por col numericas y categoricas
    - dtypes por columna
    - shape
    - info
    - total de nulos
    - total de duplicados)"""
       
    for x, df in enumerate(lista):
        print(f"------Dataframe{x}------")
        print("-------Descripción numéricas:---------")
        print(df.describe())
        print("-------Descripción categoricas:---------")
        print(df.describe(include="O"))
        print("------Tipos:---------")
        print(df.dtypes)
        print("------Forma del DataFrame:------")
        print(df.shape)
        print("------Información:---------")
        print(df.info())
        print("------Nulos:---------")
        print(df.isnull().sum())
        print("------Duplicados:---------")
        print(df.duplicated().sum())
    
exploracion(lista_df)
```

2.1 Limpiar los datos: Manejar valores nulos, eliminar duplicados si los hay, corregir errores tipográficos, etc.

```python
# Le añadimos un nuevo indice a productos_df para que la primera columna no sea el indice
cambio_productos = productos_df.reset_index()
```

```python
cambio_productos.head()
```

```python
#las cabeceras estan desplazadas a la derecha, las movemos y a la ultima columna le añadiremos el nombre extra
# Definir las nuevas cabeceras	
nuevas_cabeceras = ['ID', 'Nombre_Producto', 'Categoría', "Precio", "Origen", "Descripción", "Extra"]

# Asignar las nuevas cabeceras al DataFrame
cambio_productos.columns = nuevas_cabeceras
cambio_productos.head(2)
```

```python
# La columna extra es parte de Descripción, las unimos en una nueva
cambio_productos['Descripcion'] = cambio_productos.apply(lambda row: str(row['Descripción']) + "," + str(row['Extra']), axis=1)

cambio_productos.head(2)
```

```python
cambio_productos.drop(columns = ["Descripción", "Extra"], inplace=True)
cambio_productos.head(2)
```

```python
# Corregir nombres de las cabeceras de los dataframe
lista_df2 = [ventas_df, cambio_productos, clientes_df]

for df in lista_df2:
    df.columns = [col.capitalize() for col in df.columns]
```

```python
ventas_df.head()
```

```python
cambio_productos.head()
```

```python
clientes_df.head()
```

```python
# No hay presencia de duplicados en ninguno de los dataframes general.
```

```python
# Manejo de nulos:
# Ventas: No hay nulos
# Productos: presencia de 1 no convertimos a desconocido
cambio_productos["Descripcion"]= cambio_productos["Descripcion"].fillna("Desconocido")
```

```python
# Clientes: 
# calculamos el % de nulos por columna
df_nulos = pd.DataFrame((clientes_df.isnull().sum() / clientes_df.shape[0]) * 100, columns = ["%_nulos"])
# filtramos el DataFrame para quedarnos solo con aquellas columnas que tengan nulos
df_nulos[df_nulos["%_nulos"] > 0]
```

```python
nulos_esta_cat = clientes_df[clientes_df.columns[clientes_df.isnull().any()]].select_dtypes(include = "O").columns
print("Las columnas categóricas que tienen nulos son : \n ")
print(nulos_esta_cat)
```

2.2 Realizar la integración de datos: Unir los conjuntos de datos apropiados para obtener una tabla única que contenga información de ventas junto con detalles de productos y clientes.

```python
ventas_df.shape
```

```python
cambio_productos.shape
```

```python
clientes_df.shape
```

```python
# Unir ventas_df con cambio_productos
merged_df = pd.merge(ventas_df, cambio_productos, left_on="Id_producto", right_on="Id", how='left')
merged_df.shape
```

```python
# Unir el resultado anterior con clientes_df
merged_df = pd.merge(merged_df, clientes_df, left_on='Id_cliente', right_on='Id', how='right')

# Mostrar el DataFrame resultante
merged_df.shape
```

```python
merged_df.sample(10)
```
